#### 广播机制

相同形状的两个张量可以执行按元素操作。对于 **不同形状的张量** ，可以通过 **广播机制（broadcasting mechanism）** 执行按元素操作。

工作机制如下：
1. 通过适当复制元素来拓展一个或两个数组，以便在转换之后，两个张量具有相同的形状；
2. 对生成的数组执行按元素操作。

在大多数情况下，我们将沿着数组中长度为1的轴进行广播。

```python
a = torch.arange(3).reshape((3, 1)) 
b = torch.arange(2).reshape((1, 2))
a, b
```

```python
a+b  #广播机制，矩阵a将复制列，矩阵b将复制行，然后再按元素相加
```

#### 节省内存

```python
before = id(Y)
Y = Y + X
id(Y) == before #张量Y的内存地址发生变化
```

说明一些操作会为其结果分配新内存。可以使用 **切片表示法** 将操作的结果分配给先前分配的数组，以实现 **原地操作**。

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

#### 处理缺失值

在pandas中，NaN项代表缺失值。

为了处理缺失的数据，典型的方法包括 **插值法** 和 **删除法**：
- 插值法：用一个替代值弥补缺失值
- 删除法：直接忽略缺失值

插值法：

```python
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
# 分离输入特征和输出标签， iloc用于基于位置选择数据

# inputs = inputs.fillna(inputs.mean())将会报错（报错信息如下）
# mean()函数计算每列的均值，fillna()函数将缺失值替换为均值

inputs = inputs.fillna(inputs.select_dtypes(include='number').mean())
# 这里使用select_dtypes(include='number')选择数值类型的列来计算均值，避免非数值列导致的错误
print(inputs)

inputs = pd.get_dummies(inputs, dummy_na=True)
# get_dummies()函数将分类变量转换为独热编码，dummy_na=True表示将缺失值也作为一个类别处理
# 独热编码会为每个类别创建一个新的二进制列
print(inputs)
inputs = inputs.fillna(inputs.mean())
print(inputs)
```

## 2.3 线性代数

由于具有“线性代数”和“面向人工智能的线性代数”课程的知识，仅做补充：

- 向量或轴的维度被用来表示向量或轴的长度，即向量或轴的元素数量。张量的某个轴的维数就是这个轴的长度。
- 访问张量的转置可以使用 ``.T()`` 方法。
- 复制（为复制变量赋值新内存）可以使用 ``.clone()`` 方法。


```python
A = torch.arange(20).reshape(5, 4)
B = A.clone()
A, A.T, B, id(A), id(B)
```

- 两个矩阵的按元素乘法称为Hadamard积（Hadamard product）（数学符号⊙）
- 将张量乘以或加上一个标量不会改变张量的形状，其中张量的每个元素都将与标量相加或相乘。
- 使用 ``.sum()`` 可以表示任意形状张量的元素和。

```python
A.shape, A.sum()
```

- 在 ``.sum()`` 中指定 ``axis`` 沿某一维度降维以生成输出向量，该维度消失。 **每一个[]对之间是一个维度的元素**。类似地， ``.mean()`` 也可以通过指定 ``axis`` 求某一维度的平均值。

```python
A_sum_axis0 = A.sum(axis=0)
A_sum_axis0, A_sum_axis0.shape
A_sum_axis1 = A.sum(axis=1)
A_sum_axis1, A_sum_axis1.shape
```

- 若需要保持轴数不变，需在函数中额外给定 ``keepdims=true`` 。

```python
sum_A = A.sum(axis=1, keepdims=True)
sum_A
```

- 如果我们想沿某个轴计算A元素的累积总和，比如 ``axis=0`` （按行计算），可以调用 ``.cumsum()`` 函数。此函数不会沿任何轴降低输入张量的维度。

```python
A.cumsum(axis=0)
```

- 向量的点积使用 ``torch.dot(x, y)`` 执行，效果同 ``torch.sum(x * y)``。
- 矩阵-向量积使用 ``torch.mv(A, x)`` 执行。
- 矩阵-矩阵积使用 ``torch.mm(A, B)`` 执行，即矩阵乘法。
- 使用 ``torch.norm(x)`` 计算 *$L_2$* 范数。
