### 1.2 各硬件设备在 Linux 中的文件名

在 Linux 中，每个设备都被当成一个文件来对待， 举例来说，SATA 接口的硬盘文件名称是 `/dev/sd[a-d]`，即 `/dev/sda`、`/dev/sdb` 等

下面列出几个常见的设备与在 linux 中的文件名

| 设备               | 在 linux 中的文件名                                                 |
|:---------------- |:------------------------------------------------------------- |
| SCS/SATA/USB 硬盘机 | `/dev/sd[a-p]`                                                |
| USB 闪存盘          | `/dev/sd[a-p]` (同上)                                           |
| Virtl/0界面        | `/dev/vd[a-p]` 用于虚拟机内                                         |
| 软盘机              | `/dev/fd/[0-7]`                                               |
| 打印机              | `/dev/lp[0-2]` 25 针打印机、`/dev/usb/lp[0-15]` USB 打印机            |
| 鼠标               | `/dev/input/mouse` 通用、`/dev/psaux` PS/2 界面、`/dev/mouse` 当前鼠标  |
| CDROM/DVDR0M     | `/dev/scd[0-1]` 通用、`/dev/sr[0-1]` 通用，Cent0s常见、`/dev/cdrom` 当前 |
| 磁带机              | 太老不记录                                                         |
| IDE 硬盘机          | `/dev/hd[a-d]` 老式系统才有                                         |

要注意的是使用云端机时，你得到的是虚拟机，为了加速，虚拟机内的磁盘使用仿真机产生，一般被命名为 `/dev/vd[a-p]`

- 按下 **enter** 键后启动命令
- 使用反斜杠 **\\** 使 **enter** 后指令持续到下一行
- 使用 `quit` 主动退出执行的命令

### 2.5 正确关机

#### shutdown 关机

该命令会通知系统内的各个程序（processes）、服务等进行关闭，shutdown 可以达成以下工作：

- 可以自由选择关机模式：关机或重启
- 可以设定关机时间：立刻、未来的一个时间
- 可以自定义关机信息：在关机前，将设置的信息广播给在线的 user
- 可以仅发出警告信息：有时候可能需要测试、或则明确告知使用者的场景下，就可以使用该功能，但是可以不关机

语法如下：

```bash
/sbin/shutdown [-krhc] [时间 [警告信息]]
```

- k：不要真的关机，只发送警告信息
- r：在将系统的服务停掉之后就重新启动（常用)
- h：将系统的服务停掉之后，立即关机（常用）
- c：取消已经在进行的 shutdown 指令内容
- 时间：指定系统关机的时间，若没有时间，则默认 1 分钟后自动进行

`ls` 指 **list** ，查看当前目录的文件。 `a` 将隐藏文件显示， `l`  显示详细信息。


```bash
ls -al
total 44
# 权限  连接 拥有者 群组 文件容量 修改日期 文档名
drwxr-x--- 6 x    x    4096 Dec 23 22:17 .
```


- 权限中，第一个位置表示文件属性，如d是目录、-是文件。后面三个 `rwx` 一组，r 代表 read，w 代表 write， x 代表 execute（可执行）。第一组表示文件拥有者的权限；第二组表示加入此群组的账户的权限；第三组表示非本人且没有加入本群组的其他账户的权限。

- 一个点 `.` 表示隐藏文件。

### 3.2 文件属性编辑

#### chgrp 改变文件所属群组

```bash
chgrp [-R] [群组] 文件或目录
```


- -R：递归变更，如果是目录，则变更该目录以及目录下的所有文件




#### chown 改变文件拥有者


```bash
chown [-R] 账户名称 文件或目录
chown [-R] 账户名称:组名 文件或目录
```


使用 `chown` 可以同时修改拥有者和群组。

#### chmod 改变文件的权限

`chmod` 指令改变权限可以使用数字和符号。前面已经提到一组rwx代表一个身份的权限，rwx各自的代表数字如下：

- r = 4
- w = 2
- x = 1

常用的权限数值如下：

- `-rw-rw-r--` 664：一般文件，可读可写无执行
- `-rwxr-x-r-x` 755：shell脚本文件，拥有者可读写执行，其他的都只能刻度可执行、不可编辑
- `-rwxr------` 740：不希望该文件被其他人看到


```bash
chmod [-R] xyz 文件或目录
chmod 740 text.txt
```


- xyz：就是刚刚的数值类型的权限范围，为 rwz 属性数值的相加
- -R：递归

字母表示：

- user：u
- group：g
- other：o
- 所有：a


```bash
chmod u/g/o/a +/-/= r/w/x 文件或目录
```

### 3.4 环境路径 $PATH

- 指令应该要放置到正确的目录下，执行才比较方便
- 当前目录「.」建议不要放到 PATH 中

### 3.6 文件与目录的权限

文件有若干的属性，如读写执行等基本权限（rwx），是否为目录（d）、文件（-）或则是链接（l）等属性，修改属性等也可通过 chgrp、chown、chmod 。

**基本九权限** ：三组 `r` ， `w` ， `x` ，分属于 所有者（user）、所属组（group）、其他用户（others）。

#### umask 控制新创建文件和目录的默认权限

#### chattr 配置文件隐藏属性

#### lsattr 显示文件隐藏属性

除了 `r` ， `w` ， `x` 外，还会出现在权限上的标志有 `s` ， `t` 等：

#### 权限 SUID

当 `s` 标志出现在文件拥有者 `x` 权限上时，就被称为 Set UID（Set User ID） ，简称 SUID 特殊权限。

当用户执行带有 SUID 位的文件时，程序会以文件所有者的权限运行，而不是执行者的权限。

#### 权限 SGID

`s` 在群组的 `x` 时称为 Set GID（Set Group ID） ，简称 SGID 特殊权限。

对于文件，执行时会以文件所属组的权限运行；对于目录，在该目录下创建的新文件会自动继承目录的组。

#### 权限 SBIT

其他用户（Others）的 `x` 权限位会显示为 `t` 或 `T` ，此时 SBIT（Sticky Bit） 已设置。

SBIT 主要用于共享目录，确保用户只能删除或修改自己创建的文件。

#### SUID、SGID、SBIT 的设定

在 3.2 文件属性设定 中讲到 `chmod` 可以使用数值来设置权限。只需要在对应的数值前增加代表以上特殊权限的数值，即可设置以上特殊权限：

- SUID：4
- SGID：2
- SBIT：1

```bash
chmod 4755 test
```

#### find 实时搜索文件

`find` 是 Linux 下功能最强大的文件搜索工具，支持实时搜索、复杂条件匹配、批量处理文件，相比 `locate` 速度较慢但更灵活精准。

```bash
find [path] [option] [action]
```

**与时间有关的参数** ：

- `mtime n` ：在 n 天前的一天内被修改过内容的文件
- `mtime +n` ：列出在 n 天之前（不含 n 本身）被修改过内容的文件
- `mtime -n` ：列出在 n 天之内（含 n 天本身）被修改过内容的文件
- `newer file` ： file 为一个存在的文件，列出比 file 还要新的文件



**与使用者或组名有关的参数** ：

- `uid n`: n 为数字，是用户的账户 ID（UID），UID 记录在 `/etc/passwd` 里面与账户名称对应的数字。
- `gid n`: n 为数字，是组名的 ID，记录在 `/etc/group` 文件中。
- `user name`: name 为使用者账户名称，如 `mrcode`。
- `group name`: name 为组名。
- `nouser`: 寻找文件的拥有者不存在于 `/etc/passwd` 的文件。
- `nogroup`: 寻找文件的拥有群组不存在于 `/etc/group` 的文件。

**与文件权限及名称有关的参数** ：

- `name filename`: 查找文件名为 `filename` 的文件。
- `size [-+]SIZE`: 查找比 `SIZE` 还要大（`+`）或则小（`-`）的文件。

  `SIZE` 支持的单位有：
  - `c`: byte
  - `k`: 1024 byte

  所以要查找比 50 KB 还要大的文件，指令为：
  ```bash
  find /home/ -size +50k
  ```

- `type TYPE`: 查找文件类型为 `TYPE` 的。主要有：
  - `f`: 一般正规文件
  - `b`, `c`: 装置文件
  - `d`: 目录
  - `l`: 连接
  - `s`: socket
  - `p`: FIFO

- `perm mode`: 查找文件权限「刚好等于」`mode` 的文件，`mode` 为类似 `chmod` 的属性。


- `perm -mode`: 查找文件权限「必须要全部包括 `mode` 的权限」的文件。

- `perm /mode`: 查找文件权限「包含任意 `mode` 的权限」的文件。




## 4 磁盘与文件系统管理

### 4.1 磁盘组成

磁盘的组成主要有：

- 圆形的磁盘（主要记录数据的部分）
- 机械手臂，与机械手臂上的磁盘读取头（可擦写磁盘上的数据）
- 主轴马达，可以转动磁盘，让机械手臂的读取头在磁盘上读写数据

磁盘上的物理组成：

- 扇区（Sector）：最小的物理存储单位，目前主要有 512 bytes 和 4k 两种格式
- 磁柱（Cylinder）：扇区组成的圆。早期的分区主要以磁柱为最小单位，现在的分区通常使用扇区为最小分区单位。

磁盘分区表：

存储在磁盘起始位置的数据结构，用于定义磁盘的分区布局，磁盘分区表主要有两种格式：

- MBR 分区表：第一个扇区中包含了主要开机区（Master boot record，MBR，占用446 bytes），第二个扇区中包含了分区表（partition table，占用64 bytes）
- GPT 分区表：除了分区数量扩充较多外，支持的磁盘容量也可以超过2 TB

在 Linux 系统中，虚拟磁盘和实体磁盘的设备文件名遵循特定的命名规则，它们的命名方式取决于磁盘类型以及是否是虚拟设备：

- 实体磁盘：`/dev/sd[a-p]` ，例如第一个磁盘文件名为 `/dev/sda`
- 虚拟磁盘：`/dev/vd[a-p]` ，如有使用到软件磁盘阵列，还有 `/dev/md[0-128]` 的磁盘文件名，使用的是 LVM 时，则为 `/dev/VGNAME/LVNAME` 等格式

无论是物理还是虚拟磁盘，分区会在设备名后追加数字，如 `/dev/sd[a-p][1-128]` 和 `/dev/vd[a-p][1-128]` 。


#### 文件身份信息 Inode

在 Linux 和类 Unix 系统中， **Inode（Index Node）** 是文件系统的核心数据结构，用于存储文件的 **元数据（metadata）** 和 **数据块的位置信息** 。每个文件或目录都有一个唯一的 Inode ，它相当于文件的“身份证”，但不包含文件名本身。

前面文件系统中了解到：

- 每个文件都会占用一个 `inode`，文件内容由 `inode` 记录指向。
- 想要读取该文件，必须要经过目录记录的文件名来指向到正确的 `inode` 号码才能读取。

也就是说，文件名只与目录有关，但是文件内容则与 `inode` 有关。

#### 链接 Link

在 Linux 和类 Unix 系统中， **链接（Link）** 是一种特殊的文件类型，用于指向另一个文件或目录。链接主要分为两种主要类型： **硬链接（Hard Link）** 和 **符号链接（Symbolic Link，软链接）** 。

- **硬链接（Hard Link）** ：
    - 特点：
        1. 硬链接与原始文件共享相同的 Inode（文件在磁盘上的元数据），因此它们实际上是同一个文件的多个名称。
        2. 删除原始文件后，硬链接仍能访问数据（因为数据未被真正删除，直到所有硬链接被移除）。

    - 创建：
    ```bash
    ln [源文件] [硬链接名]
    ```

    ```bash
    ln file.txt hardlink_to_file.txt
    ```

- **符号链接（Symbolic Link）** ：
    - 特点：
        1. 软链接是一个独立的文件，内容为目标文件的路径（类似于 Windows 快捷方式）
        2. 可以跨文件系统和链接到目录，删除原文件后，软链接会无法访问文件
        3. 软链接的权限是 777，但实际权限由目标文件决定
    - 创建：
    ```bash
    ln -s 目标文件或目录 软链接名
    ```

    ```bash
    ln -s /path/to/file.txt symlink_to_file.txt
    ```

管理链接的命令如下：

| 命令                | 作用                                |
|---------------------|-----------------------------------|
| `ln`                | 创建硬链接                        |
| `ln -s`             | 创建软链接                        |
| `ls -li`            | 查看文件 Inode 和链接数           |
| `readlink`          | 查看软链接指向的实际路径          |
| `find -type l`      | 查找当前目录下的所有软链接        |
| `stat 文件名`       | 查看文件的详细信息（包括链接数）  |

### 4.5 挂载，设置开机挂载

**挂载（Mounting）** 是Linux/Unix系统中将存储设备（如硬盘分区、CD-ROM、网络共享等）关联到文件系统目录树的过程。通过挂载，用户可以通过目录（挂载点）访问设备中的数据。

简单来说，Linux中所有存储设备都是独立的文件系统，通过挂载， **设备可以被链接到目录树中的特定为止** 。


#### mount 挂载

在Linux系统重需要使用 `mount` 指令进行挂载：

```bash
mount -a
mount [-l]
mount [-t 文件系统] LABEL='' 挂载点
mount [-t 文件系统] UUID='' 挂载点
mount [-t 文件系统] 装置文件名='' 挂载点
```


#### umount 将装置文件卸除

```bash
umount [-fn] 装置文件名或挂载点

-f：强制卸除，可用在类似网络文件系统无法读取到的情况下
-l：like卸除文件系统，比 -f 还强
-n：不更新 /etc/mtab 情况下卸除
```

#### 设置开机挂载

手动处理 `mount` 后不是永久挂载的，重启后挂载失效。

可以通过修改 `/etc/fstab` 文件来达到开机自动挂载。




**开机挂载** 是系统启动时自动执行挂载操作的过程，确保关键存储资源在系统启动后立即可用。




#### 特殊装置 loop 挂载

**Loop 设备挂载** 是一种特殊的挂载方式，它允许将 **普通文件（如 ISO 镜像、磁盘映像）** 当作 **块设备** 挂载到文件系统中。这种机制在 Linux 中通过 `loop` 设备实现，创建了一个"虚拟磁盘驱动器"的效果。

## 5 压缩、打包、备份



计算机最小的计量单位是 bits，目前使用的计算机系统中都是使用 bytes 单位来计量的，当 1 byte = 8 bits 中没有被填满时，就会出现大量的 bit 被 0 填充。

通过一些复杂的计算方式将这些没有使用到的空间去掉，来达到让文件占用空间变小的目的，就是 **压缩** 的原理。



- .Z: compress 程序压缩的文件
- .zip: zip 程序压缩的文件
- .gz: gzip 程序压缩的文件
- .bz2: bzip2 程序压缩的文件
- .xz: xz 程序压缩的文件
- .tar: tar 程序打包的数据，并没有压缩过
- .tar.gz: tar 程序打包的数据，并经过 gzip 的压缩
- .tar.bz2: 同上，经过了 bzip2 压缩
- .tar.xz: 同上，经过了 xz 的压缩


| 指令    | 指令作用                             | 基本语法                             |
|---------|--------------------------------------|--------------------------------------|
| **gzip** | 使用 LZ77 算法压缩文件，生成 `.gz` 格式 | `gzip [选项] 文件名`                 |
|         |                                      | 压缩：`gzip file.txt` → `file.txt.gz` |
|         |                                      | 解压：`gunzip file.txt.gz` 或 `gzip -d file.txt.gz` |
| **bzip2** | 使用 Burrows-Wheeler 算法压缩，生成 `.bz2` 格式 | `bzip2 [选项] 文件名`              |
|         |                                      | 压缩：`bzip2 file.txt` → `file.txt.bz2` |
|         |                                      | 解压：`bunzip2 file.txt.bz2` 或 `bzip2 -d file.txt.bz2` |
| **xz**   | 使用 LZMA2 算法压缩，生成 `.xz` 格式，压缩率最高 | `xz [选项] 文件名`                 |
|         |                                      | 压缩：`xz file.txt` → `file.txt.xz`   |
|         |                                      | 解压：`unxz file.txt.xz` 或 `xz -d file.txt.xz` |

#### tar 打包工具

Tar（Tape Archive）是 Linux/Unix 系统中用于 文件打包的核心工具，它：

- 将多个文件/目录合并为单个文件（归档）
- 保留文件属性（权限、时间戳、所有权）
- 常与压缩工具（gzip/bzip2/xz）配合使用
- 支持创建、查看、提取归档文件


```bash
打包与压缩：`tar [-z|-j|-J][cv][-f 待建立的文件名] filename`
观察文件：  `tar [-z|-j|-J][tv][-f file.tar]`
解压缩：	  `tar [-z|-j|-J][xf][-f file.tar] [-C 目录]`

特别注意：`[-z|-j|-J]` 不可同时出现在一串指令中
特殊注意：c、t、x 也不可同时出现在一串指令中
```
